```python
'''
Please refer to the following link for more instructions about using corr and cov function in pandas.

For correlation coefficient:
For series: https://pandas.pydata.org/docs/reference/api/pandas.Series.corr.html#pandas.Series.corr
For DataFrame: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html#pandas-dataframe-corr

For covariance: 
For series: https://pandas.pydata.org/docs/reference/api/pandas.Series.cov.html#pandas-series-cov
For DataFrame: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cov.html#pandas.DataFrame.cov

'''

import pandas as pd

s1 = pd.Series([2, 3, 5, 4, 6])
s2 = pd.Series([5, 8, 10, 11, 14])

print("correlation coefficient:", s1.corr(s2))
print("covariance normalized by n-1:", s1.cov(s2))     #covariance normalized by n-1
print("covariance normalized by n:", s1.cov(s2)*4/5)  #covariance normalized by n as we discussed in class
```

```python
import pandas as pd

from sklearn.datasets import fetch_california_housing
house = fetch_california_housing()

print(house.data.shape)
print(house.feature_names)

hou = pd.DataFrame(house.data, columns = house.feature_names)

hou.corr()  #calculate the pairwise correlation coefficient between attributes
```
