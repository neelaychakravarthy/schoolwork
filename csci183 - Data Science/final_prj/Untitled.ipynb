```python
import pandas as pd
```

```python
east_conf = pd.read_csv('22-23_eastconf.csv')
east_conf
```

```python
import matplotlib.pyplot as plt
plt.scatter(east_conf.EasternConference , east_conf.PSG , s=12, marker='o')
plt.xlabel('Team')
plt.ylabel('PPG')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x_train = east_conf.drop(['EasternConference','PLY'], axis=1)
y_train = east_conf['PLY']
```

```python
west_conf = pd.read_csv('22-23_westconf.csv')
west_conf
```

```python
x_test = west_conf.drop(['WesternConference', 'PLY'], axis=1)
y_test = west_conf['PLY']
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred = log_reg.predict(x_test)
```

```python
print(accuracy_score(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```

```python
log_reg.fit(x_test, y_test)
y_pred = log_reg.predict(x_train)
```

```python
print(accuracy_score(y_pred, y_train))
print(confusion_matrix(y_pred, y_train))
```
