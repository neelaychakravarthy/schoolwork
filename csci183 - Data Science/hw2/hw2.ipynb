```python
#Question 7
import scipy as sp

lr = sp.stats.linregress(housing_numerical['median_income'], housing_numerical['median_house_value'])
df = pd.DataFrame({'Feature':['median_income'], 'Slope':[lr.slope], 'Intercept':[lr.intercept]})
lr
```

```python
plt.scatter(housing_numerical['median_house_value'] , housing_numerical['median_income'] , s=12, marker='o')
plt.axline(xy1=(0, lr.intercept), slope=lr.slope)
plt.xlabel('house value')
plt.ylabel('median_income')
plt.show()
```

```python
lr = sp.stats.linregress(kc_house_data_numerical['sqft_living'], kc_house_data_numerical['price'])
df.loc[len(df.index)] = ['sqft_living', lr.slope, lr.intercept] 
lr
```

```python
plt.scatter(kc_house_data_numerical['price'] , kc_house_data_numerical['sqft_living'] , s=12, marker='o')
plt.axline(xy1=(0, lr.intercept), slope=lr.slope)
plt.xlabel('price')
plt.ylabel('sqft_living')
plt.show()
```

```python
lr = sp.stats.linregress(kc_house_data_numerical['grade'], kc_house_data_numerical['price'])
df.loc[len(df.index)] = ['grade', lr.slope, lr.intercept] 
lr
```
