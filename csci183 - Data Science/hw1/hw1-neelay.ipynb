```python
#Question 1 - Python Basics

def isPrime(n):

    #base case
    if(n <= 1):
        return False
    for i in range(2, n):
        if n%i == 0:
            return False
    return True

#given number
given_number = 44
print(given_number)
for i in range(2, given_number+1):
    if isPrime(i):
        print(i)
```

```python
#Question 2 - Python Basics

def largestAdjacentProduct(list_of_numbers):
    max_prod = 0
    for i in range(len(list_of_numbers) - 1):
        temp = list_of_numbers[i] * list_of_numbers[i+1]
        if(temp > max_prod):
            max_prod = temp
    print(max_prod)

#given list of numbers
given_list_of_numbers = [1, 3, 2, 4, 67, 3, 6, 7, 3, 5, 23, 5, 6, 2, 4, 45]
print(given_list_of_numbers)
largestAdjacentProduct(given_list_of_numbers)
```

```python
#Question 3 - Python Basics

import string

def isValidPwd(pwd):
    oneNumber = False
    oneLowerCase = False
    oneUpperCase = False
    oneSpecial = False
    if len(pwd) not in range(6, 21):
        print('Invalid Password: Length')
    for i in range(len(pwd)):
        if pwd[i].isnumeric():
            oneNumber = True
        if pwd[i].isupper():
            oneUpperCase = True
        if pwd[i].islower():
            oneLowerCase = True
        if pwd[i] in set(string.punctuation):
            oneSpecial = True
    if oneNumber and oneLowerCase and oneUpperCase and oneSpecial:
        print('Valid Password!')
        return True
    else:
        print('Invalid Password: A valid password - ')
        if not oneNumber:
            print('     Should have at least one number.')
        if not oneLowerCase or not oneUpperCase:
            print('     Should have at least one uppercase and one lowercase character.')
        if not oneSpecial:
            print('     Should have at least one special symbol.')
        return False
    
given_pwd = 'sadfTY23^'
isValidPwd(given_pwd)
```

```python
#Question 6 - Python for Data Science
def isLength(variable, n): 
    if len(variable) == n: 
        return True 
    return False

def getNLengthWords(): 
    file = open("Data Science.txt") 
    text = file.read() 
    print(text)
    words = text.split()

    for i in range (len(words)):
        words[i] = words[i].lower()

    print("Enter length: ")
    n = int(input())
    words = filter(lambda seq: isLength(seq, n), words)
    words = list(set(list(words)))
    
    print(words)
getNLengthWords()
```
